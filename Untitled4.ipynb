```python
df_BKK = pd.read_csv("data/db/bkk-data.csv", sep=";", index_col=0, dtype={'TimeStamp': 'string', 'Terminal': 'string', 'ScanningPos': 'string', 'Destination':'string', 'TravelDate':'string', 'Evaluation':'string', 'STD':'string', 'DocumentType':'string'})
df_Coronazahlen = pd.read_csv("data/db/new_cases.csv", sep=",")
df_Flughaefen = pd.read_csv("data/db/airports.csv", sep=",")
df_Standortestammdaten = pd.read_csv("data/db/GDP.csv", sep=",")

df_BKK.to_sql(con=db_connection, name='BKK', if_exists='replace')
df_Coronazahlen.to_sql(con=db_connection, name='Coronazahlen', if_exists='replace')
df_Flughaefen.to_sql(con=db_connection, name='Flughaefen', if_exists='replace')
df_Standortestammdaten.to_sql(con=db_connection, name='Standortestammdaten', if_exists='replace')

```
